```python
import pandas as pd
import numpy as np

granify_data = pd.read_csv(r"C:\Users\level\Downloads\granifyproject-main\granifyproject-main\BA Candidate Assignment 1 Files [EXTERNAL]\granify_orders.csv")
crisskross_data = pd.read_csv(r"C:\Users\level\Downloads\granifyproject-main\granifyproject-main\BA Candidate Assignment 1 Files [EXTERNAL]\crisskross_orders.csv")

crisskross_data = crisskross_data.rename(columns={"Revenue" : "Revenue_ck", "Segment" : "Segment_ck"})

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

print(granify_data.head())
print(crisskross_data.head())
```

```python
combined_data = granify_data.join(crisskross_data.set_index("Transaction ID"), on="order_id")

print(combined_data.head())
```

```python
#Flags rows with discrepancies in revenue
combined_data["revenue_flag"] = np.where(combined_data["subtotal"] != combined_data["Revenue_ck"], 1, 0)

print(combined_data.head())
```

```python
#This is to check to verify that 1 denotes a baseline group member. If I sum the column and the sum is roughly 10% of the number
#of entries, then 1 does in fact denote the baseline group.
print(np.sum(combined_data["test_group"]))

combined_data.info()
```

And we can see that 12,499 rows have a "test_group" value of 1, which is roughly 10% of the 127,199 total rows.

```python
#Numeric version of Segement_CK so that it can easily be compared with test_group
combined_data["segment_num"] = np.where(combined_data["Segment_ck"] == "granify", 0, 1)

print(combined_data.head())

#Flags rows with discrepancies in orders
combined_data["order_flag"] = np.where(combined_data["test_group"] != combined_data["segment_num"], 1, 0)
```

```python
order_discrepancies = np.sum(combined_data["order_flag"])
revenue_discrepancies = np.sum(combined_data["revenue_flag"])
total_orders = len(combined_data.index)

combined_data["discrepancy_flag"] = np.where((combined_data["revenue_flag"] == 1) | (combined_data["order_flag"] == 1), 1, 0)

any_discrepancy = np.sum(combined_data["discrepancy_flag"])

print("The number of rows with revenue discrepancies is", revenue_discrepancies, "or", "{:%}".format(revenue_discrepancies/total_orders))
print("The number of rows with order discrepancies is", order_discrepancies, "{:%}".format(order_discrepancies/total_orders))
print("The number of rows with at least one discrepancy is", any_discrepancy, "{:%}".format(any_discrepancy/total_orders))
```

Ok, so there are discrepancies about 4% of the time on both revenue and orders. While not terrible, that's not great either. As per the instructions, Granify prefers the delta to be under 3%.

First I'll investigate the discrepancies between order type. Below I'm playing around with the data to see if there are any patterns in the flagged orders versus all orders.

```python
#Creates a dataframe that contains only rows with order discrepancies
flagged_orders = combined_data[combined_data["order_flag"] == 1]
non_flagged_orders = combined_data[combined_data["order_flag"] == 0]

print(flagged_orders.info())
```

```python
print("{:%}".format(np.sum(flagged_orders["revenue_flag"])/len(flagged_orders.index)), "of order discrepancies also had a revenue discrpency.")
```

```python
print(flagged_orders["os"].value_counts())
print("\n", flagged_orders["browser"].value_counts())
print("\n", flagged_orders["device_category"].value_counts())
print("\n", flagged_orders["test_group"].value_counts())
print("\n", flagged_orders["campaign_id"].value_counts())
```

```python
func1 = lambda x: 100*x.count()/flagged_orders.shape[0]
func2 = lambda x: 100*x.count()/non_flagged_orders.shape[0]

flagged_os_check = pd.pivot_table(flagged_orders, index =['os', "os_version"], values="order_flag", aggfunc=func1)
combined_os_check = pd.pivot_table(non_flagged_orders, index =['os', "os_version"], values="order_flag", aggfunc=func2)

os_check = flagged_os_check.join(combined_os_check, lsuffix="_flagged", rsuffix="_nf")
os_check["diff"] = os_check["order_flag_flagged"] - os_check["order_flag_nf"]
display(os_check.sort_values("diff"))


```

```python
flagged_browser_check = pd.pivot_table(flagged_orders, index =['browser', "browser_version"], values="order_flag", aggfunc=func1)
combined_browser_check = pd.pivot_table(non_flagged_orders, index =['browser', "browser_version"], values="order_flag", aggfunc=func2)

browser_check = flagged_browser_check.join(combined_browser_check, lsuffix="_flagged", rsuffix="_nf")
browser_check["diff"] = browser_check["order_flag_flagged"] - browser_check["order_flag_nf"]
display(browser_check.sort_values("diff"))

browser_check.to_csv(r"C:\Users\level\Downloads\granifyproject-main\granifyproject-main\BA Candidate Assignment 1 Files [EXTERNAL]\order_discrepancy_browser.csv")
```

```python
flagged_traffic_type_orders = pd.pivot_table(flagged_orders, index =["traffic_type"], values="order_flag", aggfunc="count")
not_flagged_traffic_type_orders =pd.pivot_table(non_flagged_orders, index =["traffic_type"], values="order_flag", aggfunc="count")

display(pd.pivot_table(flagged_orders, index =["order_flag"], values="date", aggfunc=[max, min]))
display(pd.pivot_table(non_flagged_orders, index =["order_flag"], values="date", aggfunc=[max, min]))

traffic_type_check_orders = flagged_traffic_type_orders.join(not_flagged_traffic_type_orders, lsuffix="_fl", rsuffix="_not") 

display(traffic_type_check_orders)

traffic_type_check_orders.to_csv(r"C:\Users\level\Downloads\granifyproject-main\granifyproject-main\BA Candidate Assignment 1 Files [EXTERNAL]\order_discrepancy_traffic_type.csv")
```
